```python
plt.figure(figsize=(8,5))
plt.hist(df['Price'],bins=10)
plt.title("Distribution Of Restaurant Prices")
plt.xlabel("Price")
plt.ylabel("Number of Restaurants")
plt.show()
```

Plotting of the Distribution of Restaurant Prices

```python
plt.figure(figsize=(8,5))
plt.hist(df['Delivery time'],bins=10)
plt.title("Distribution Of Delivery Time")
plt.xlabel('Delivery Time(In Mintutes)')
plt.ylabel('Number of Restaurants')
plt.show()
```

 Plotting of Distribution Of Delivery Time

```python
city_count=df["City"].value_counts().head(10)
city_count.plot(kind='bar',figsize=(10,5))
plt.title("Top 10 Cities By Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Restaurant Count")
plt.show()
```

Top 10 Cities by Number Of Restaurants

```python
plt.scatter(df['Price'],df['Avg ratings'])
plt.title("Price vs Average Ratings")
plt.xlabel("Price")
plt.ylabel("Average Ratings")
plt.show()
```

Plotting Shows the Price vs Average Ratings

```python
df['Avg ratings'].head(20).plot(kind='line')
plt.title("Average Ratings")
plt.xlabel("Restaurants")
plt.ylabel("Rating")
plt.show()
```

Plotting of Average Ratings Using A line Chart
